# Pandas
판다스(Pandas)는 데이터 분석과 관련된 다양한 기능을 제공하는 파이썬 라이브러리이다.

### A. 데이터 만들기

* 1. 판다스 라이브러리 불러오기

```python
import pandas as pd
```

* 2. 데이터 프레임 만들기

```python
df = pd.DataFrame(          # df라는 변수에 데이터 프레임을 담아 준다.
    {"a" : [4, 5, 6, 7],    # 열 이름을 지정해 주고 시리즈 형태로 데이터를 저장한다.
     "b" : [8, 9, 10, 11],
     "c" : [12, 13, 14, 15]},
    index=[1, 2, 3, 4])     # 인덱스는 1, 2, 3, 4로 정해준다.
```

* 3. 데이터 프레임 출력하기

```python
df = pd.DataFrame(
    [[4, 8, 12],
     [5, 6, 13],
     [6, 10, 14],
     [7, 11, 15]],
    index=[1, 2, 3, 4],
    columns=['a', 'b', 'c']) # 열 이름을 따로 정해 줄 수 있다.
```

### 행과 열 변환하기

### M. 데이터 병합하기

```python
# 실습을 위한 데이터 프레임 만들기 1
adf = pd.DataFrame({"x1" : ["A", "B", "C"],
                    "x2" : [1, 2, 3]})
```

```python
adf
```

```python
# 데이터 프레임 만들기 2
bdf = pd.DataFrame({"x1" : ["A", "B", "D"],
                    "x3" : ["T", "F", "T"]})
```

```python
bdf
```

```python
# 데이터 프레임 만들기 3
cdf = pd.DataFrame({"x1" : ["B", "C", "D"],
                    "x2": [2, 3, 4]})
```

```python
cdf
```

* 70. 왼쪽 열을 축으로 병합하기

```python
# x1을 키로 해서 병합, 왼쪽(adf)을 기준으로
# 왼쪽의 adf에는 D가 없으므로 해당 값은 NaN으로 변환한다.
pd.merge(adf, bdf, how='left', on='x1')
```

* 71, 오른쪽 열을 축으로 병합하기

```python
pd.merge(adf, bdf, how='right', on='x1')
```

- 72. 공통 값만 병합하기

```python
pd.merge(adf, bdf, how='inner', on='x1')
```

- 73. 모든 값을 병합하기

```python
pd.merge(adf, bdf, how='outer', on='x1')
```

- 74. 특정한 열을 비교해서 공통 값이 존재하는 경우만 가져오기

```python
# adf와 bdf의 특정한 열을 비교해서 공통 값이 존재하는 경우만 가져온다.
# adf.x1열과 bdf.x1열은 A와 B가 같으므로 adf의 해당 값만 출력한다.
adf[adf.x1.isin(bdf.x1)]
```

* 75. 공통 값이 존재하는 경우 해당 값을 제외하고 병합하기

```python
adf[~adf.x1.isin(bdf.x1)]
```

* 76. 공통 값이 있는 것만 병합하기

```python
pd.merge(adf, cdf)
```

* 77. 모두 병합하기

```python
pd.merge(adf, cdf, how='outer')
```

* 78. 어디서 병합되었는지 표시하기

```python
pd.merge(adf, cdf, how='outer', indicator=True)
```

* 79. 원하는 병합만 남기기

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"')
```

* 80. merge 칼럼 없애기

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"').drop(columns=['_merge'])
```
