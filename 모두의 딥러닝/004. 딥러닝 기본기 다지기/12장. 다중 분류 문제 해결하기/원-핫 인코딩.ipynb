```python
import pandas as pd

# 깃허브에 준비된 데이터를 가져온다.
!git clone https://github.com/taehojo/data.git

# 아이리스 데이터를 불러온다.
df = pd.read_csv('./data/iris3.csv')

df.head()
```

불러온 데이터 프레임을 X와 y로 나눈다.

```python
X = df.iloc[:, 0:4:]
y = df.iloc[:, 4]
```

X와 y의 첫 다섯 줄을 출력한다.

```python
print(X[:5], '\n')
print(y[:5])
```

y의 값이 숫자가 아닌 문자이다. 딥러닝에서는 계산을 위해 문자를 모두 숫자형으로 바꾸어 주어야 한다.  
여러 개의 값으로 된 문자열을 0과 1로만 이루어진 형태로 만들어 주는 과정을 원-핫 인코딩(one-hot encoding)이라고한다. 원-핫 인코딩은 판다스가 제공하는 get_dummies() 함수를 사용한다.

```python
# 원-핫 인코딩 처리를 한다.
y = pd.get_dummies(y)

# 원-핫 인코딩 결과를 확인한다.
print(y[0:5])
```
