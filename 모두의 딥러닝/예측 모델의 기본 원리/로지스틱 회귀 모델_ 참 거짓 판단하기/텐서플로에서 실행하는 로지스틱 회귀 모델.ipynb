```python
-import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential() # 딥러닝의 한 층 한 층을 add()라는 함수를 사용해 간단히 추가시켜 준다.
model.add(Dense(1, input_dim=1, activation='sigmoid')) # 내부의 층을 만들 수 있다.

# 교차 엔트로피 오차 함수를 이용하기 위해 'binary_crossentropy'로 설정한다.
model.compile(optimizer='sgd', loss='binary_crossentropy')
model.fit(x, y, epochs=5000)

# 그래프로 확인해본다.
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

# 임의의 학습 시간을 집어넣어 합격 예상 확률을 예측해본다.
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%이다." % (hour, prediction * 100))
```
