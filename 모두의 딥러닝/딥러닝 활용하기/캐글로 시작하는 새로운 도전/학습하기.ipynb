케라스의 load_model과 판다스를 불러온다.

```python
from tensorflow.keras.models import load_model
import pandas as pd
```

해당 테스트셋을 불러온다.

```python
%cd /content/drive/MyDrive/Colab Notebooks/모두의 딥러닝/딥러닝 활용하기/캐글로 시작하는 새로운 도전
```

```python
kaggle_test = pd.read_csv("./data/test.csv")
```

테스트셋의 속성은 학습셋과 동일한 상태로 변형되어야 해당 모델을 적용할 수 있다. 이를 위해 학습셋과 동일하게 전처리되어야 한다. 먼저 카테고리형 변수를 0과 1로 이루어진 변수로 바꾸어준다.

```python
kaggle_test = pd.get_dummies(kaggle_test)
```

결측치를 전체 칼럼의 평균으로 대체해 채워준다.

```python
kaggle_test = kaggle_test.fillna(kaggle_test.mean())
kaggle_test
```

학습에 사용된 열을 K_test로 저장한다.

```python
cols_kaggle = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF']
K_test = kaggle_test[cols_kaggle]
```

만든 모델을 불러온다.

```python
model = load_model("./data/model/Ch15-house.hdf5")
```

model.predict()를 이용해 불러온 모델에 조금 전 만든 K_test를 적용하고 예측 값을 만든다.

```python
ids = [] # ID와 예측 값이 들어갈 빈 리스트를 만든다.
Y_prediction = model.predict(K_test).flatten()
for i in range(len(K_test)):
    id = kaggle_test['Id'][i]
    prediction = Y_prediction[i]
    ids.append([id, prediction])
```

현재 시간을 이용해 파일명을 만들어 저장한다.

```python
import time

timestr = time.strftime("%Y%m%d-%H%M%S")
filename = str(timestr) # 파일명을 연월일-시분초로 정한다.
outdir = './'           # 파일이 저장될 위치를 지정한다.
```

만들어진 실행 번호와 예측 값을 새로운 데이터 프레임에 넣고 이를 csv 파일로 저장한다.

```python
df = pd.DataFrame(ids, columns=["Id", "SalePrice"])
df.to_csv(str(outdir + filename + '_submission.csv'), index=False)
```

실습| 캐글에 제출할 결과 만들기

```python
from tensorflow.keras.models import load_model

import pandas as pd
import time

# 캐글에서 내려받은 테스트셋을 불러온다.
kaggle_test = pd.read_csv("./data/test.csv")

# 카테고리형 변수를 0과 1로 이루어진 변수로 바꾼다.
kaggle_test = pd.get_dummies(kaggle_test)

# 결측치를 전체 칼럼의 펻균으로 대체해 채워 준다.
kaggle_test = kaggle_test.fillna(kaggle_test.mean())

# 집 값을 제외한 나머지 열을 저장한다.
cols_kaggle = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF']
K_test = kaggle_test[cols_kaggle]

# 앞서 만든 모델을 불러온다.
model = load_model("./data/model/Ch15-house.hdf5")

# ID와 예측 값이 들어갈 빈 리스트를 만든다.
ids = []

# 불러온 모델에 K_test를 적용하고 예측 값을 만든다.
Y_prediction = model.predict(K_test).flatten()
for i in range(len(K_test)):
    id = kaggle_test['Id'][i]
    predition = Y_prediction[i]
    ids.append([id, prediction])

# 테스트 결과의 저장 환경을 설정한다.
timestr = time.strftime("%Y%m%d-%H%M%S")
filename = str(timestr) # 파일명을 연월일-시분초로 정한다.
outdir = './'           # 파일이 저장될 위치를 지정한다.

# Id와 집 값을 csv 파일로 저장한다.
df = pd.DataFrame(ids, columns=["Id", "SalePrice"])
df.to_csv(str(outdir + filename + '_submission.csv'), index=False)
```
