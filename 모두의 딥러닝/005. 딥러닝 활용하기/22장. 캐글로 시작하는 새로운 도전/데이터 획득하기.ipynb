1. 해당 경진대회에 접속을 완료하면 대회에 대한 내용을 숙지하고 Data를 클릭해 데이터에 접근을 한다.  
2. 데이터 화면이 나오면 I understand and agree를 클릭해 데이터를 내려받을 준비를 한다.  
3. 데이터 화면이 바뀌고 해당 데이터에 대한 설명이 나온다. Download All을 클릭한다.
4. 내려받은 데이터를 확인한다.

data_description.txt: 내려받은 데이터의 각 속성이 무엇을 의미하는지 설명  
train.csv: 집 값과 해당 집이 어떤 속성을 가졌는지 정리된 파일이다.  
test.csv: train.csv 파일을 이용해 학습한 결과를 테스트하기 위한 데이터이다. 맨 마지막 값은 빠져 있다. 이 항목을 예측하는 것이 우리의 과제이기 때문이다.  
sample_submission.csv: Id와 SalePrice 두 개의 열만 존재하는 파일이다. 각 Id별로 우리가 예측한 SalePrice를 채워 넣어 캐글에 업로드하면 된다.
