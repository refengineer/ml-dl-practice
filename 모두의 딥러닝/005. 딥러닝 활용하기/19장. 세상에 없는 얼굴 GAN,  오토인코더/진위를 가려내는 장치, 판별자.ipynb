판별자(discriminator)는 생성자에서 넘어온 이미지가 가짜인지 진짜인지를 판별해 주는 장치이다.

판별자가 얻은 가중치는 판별자가 자신이 학습하는 데 쓰이는 것이 아니라 생성자로 넘겨주어 생성자가 업데이트된 이미지를 만들도록 해야 한다. 따라서 판별자를 만들 때는 가중치를 저장하는 학습 기능을 꺼 주어야 한다.

모든 과정을 코드로 정리해 보면 다음과 같다.



```
# 모델 이름을 dicriminator로 정하고 Sequential() 함수를 호출한다.
discriminator = Sequential()
discriminator.add(Conv2D(64, kernel_size=5, strides=2, input_shape=(28,28,1), padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Conv2D(128, kernel_size=5, strides=2, padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Flatten())
discriminator.add(Dense(1, activation='sigmoid'))
discriminator.compile(loss='binary_crossentropy', optimizer='adam')
discriminator.trainable = False
```



노드의 수는 각각 64개, 128개로 정하고, 커널 크기는 5로 설정해 5x5 커널이 사용된다.
```
discriminator.add(Conv2D(64, kernel_size=5, strides=2, input_shape=(28,28,1), padding="same"))
discriminator.add(Conv2D(128, kernel_size=5, strides=2, padding="same"))
```
strides는 커널 원도를 몇 칸씩 이동시킬지 정하는 옵션이다. 특별한 설정이 없으면 커널 윈도는 한 칸씩 움직인다.  
strides를 써서 커널 윈도를 여러 칸 움직이게 하는 이유는 가로세로 크기가 더 줄여들어 새로운 특징을 뽑아 주는 효과가 생기기 때문이다.  
판별자는 진짜와 가짜만 구분하면 되기 때문에 strides나 드롭아웃 등 차원을 줄여주는 기능을 적극적으로 사용하면서 컨볼루션 신경망 본래의 목적을 달성하면 된다.


활성화 함수로 LeakyReLU() 함수를 사용한 것을 보여준다.

```
discriminator.add(Activation(LeakyReLU(0.2)))
```



가로x세로의 2차원으로 진행된 과정을 1차원으로 바꾸어주는 Flatten() 함수와 마지막 활성화 함수로 sigmoid() 함수를 사용하는 과정이다. 판별의 결과가 진짜(1) 혹은 거짓(0), 둘 중에 하나가 되어야 하므로 sigmoid() 함수를 썼다.

```
discriminator.add(Flatten())
discriminator.add(Dense(1, activation='sigmoid'))
```




이진 로스 함수(binary_crossentropy)와 최적화 함수(adam)를 써서 판별자에 필요한 준비를 마무리한다.
```
discriminator.compile(loss='binary_crossentropy', optimizer='adam')
```



판별이 끝나고 나면 판별자가 자신이 학습되지 않게끔 학습 기능을 꺼 준다. discriminator.trainable = False라는 옵션으로 이를 설정할 수 있다.

```
discriminator.trainable = False
```



````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Activation, LeakyReLU, Conv2D

# 모델 이름을 dicriminator로 정하고 Sequential() 함수를 호출한다.
discriminator = Sequential()
discriminator.add(Conv2D(64, kernel_size=5, strides=2, input_shape=(28,28,1), padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Conv2D(128, kernel_size=5, strides=2, padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Flatten())
discriminator.add(Dense(1, activation='sigmoid'))
discriminator.compile(loss='binary_crossentropy', optimizer='adam')
discriminator.trainable = False
discriminator.summary()
````
