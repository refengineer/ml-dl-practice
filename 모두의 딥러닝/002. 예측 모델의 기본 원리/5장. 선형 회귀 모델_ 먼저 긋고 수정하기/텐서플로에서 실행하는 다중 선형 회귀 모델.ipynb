# 실습| 텐서플로에서 실행하는 다중 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수들을 불러온다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])

model = Sequential() # 딥러닝의 한 층 한 층을 add()라는 함수를 사용해 간단히 추가시켜 준다.

# 입력 변수가 두 개(학습 시간, 과외 시간)이므로 input_dim에 2를 입력한다.
model.add(Dense(1, input_dim=2, activation='linear')) # model.add()는 내부의 층을 만들 수 있다.
model.compile(optimizer='sgd', loss='mse') # 만든 model의 설정을 그대로 실행하라는 의미이다.

model.fit(x, y, epochs=2000) # 딥러닝의 기본 방식은 층들이 한 번만 통과하는 것이 아니라 위아래로 여러 차례 오가며 최적의 모델을 찾는 것이다.
                            # 몇 번을 오갈 것인지, 그리고 한 번 오갈 때 몇 개의 데이터를 사용할 것인지 정하는 함수이다.

# 임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트해본다.
hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점이다." % (hour, private_class, prediction))
```
