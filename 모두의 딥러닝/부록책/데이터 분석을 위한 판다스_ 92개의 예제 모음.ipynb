# I. 새로운 열 만들기

```python
# 새로운 열 만들기 실습을 위한 데이터 프레임 만들기
df = pd.DataFrame(
    {"a" : [4, 5, 6, 7],
     "b" : [8, 9, 10, 11],
     "c" : [12, 13, 14, 15]},
    index = pd.MultiIndex.from_tuples(              # 인덱스를 튜플로 지정합니다.
        [('d', 1), ('d', 2), ('e', 1), ('e', 2)],
        names=['n', 'v']                            # 인덱스 이름을 지정합니다.
    )
)
df
```

```python
df['sum'] = df['a'] + df['b'] + df['c']
df
```

```python
# a열, b열, c열의 값을 모두 곱해 d열을 만들어 줍니다.
df = df.assign(multiply=lambda df: df['a']*df['b']*df['c'])
df
```

```python
# a열을 두 개로 나누어 각각 새롭게 레이블을 만들라는 의미입니다.
df['qcut'] = pd.qcut(df['a'], 2, labels=["600이하", "600이상"])
df
```

```python
# a열에서 5이하는 모두 5로, 6이상은 모두 6으로 변환합니다.
df['clip'] = df['a'].clip(lower=5, upper=6)
df
```

```python
df.max(axis=0) # axis=0은 행과 행 비교, axis=1은 열과 열 비교
```

```python
df.min(axis=0) # axis=0은 행과 행 비교, axis=1은 열과 열 비교
```

# J. 행과 열 변환하기

```python
# 열을 행으로, 행을 열로 변환하기

# 실습을 위해 새로운 데이터를 만듭니다.
df = pd.DataFrame({'A' : {0:'a', 1:'b', 2:'c'},
                   'B' : {0:1, 1:3, 2:5},
                   'C' : {0:2, 1:4, 2:6}})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B']) # A열만 그대로, B열은 행으로 이동
```

```python
# A열만 그대로, B열과 C열은 행으로 이동시킵니다.
df_melt = pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
df_melt
```

```python

# A열을 새로운 인덱스로 만들고, B열과 C열은 이에 따라 정리합니다.
df_pivot = df_melt.pivot(index='A', columns='variable', values='value')
df_pivot
```

```python
df_origin = df_pivot.reset_index() # 인덱스를 리셋합니다.
df_origin
```

```python
df_origin.columns.name = None # 인덱스 열의 이름을 초기화합니다.
df_origin
```

# M. 데이터 병합하기

```python
# 실습을 위한 데이터 프레임 만들기 1
adf = pd.DataFrame({"x1" : ["A", "B", "C"],
                    "x2" : [1, 2, 3]})
adf
```

```python
# 데이터 프레임 만들기 2
bdf = pd.DataFrame({"x1" : ["A", "B", "D"],
                    "x3" : ["T", "F", "T"]})
bdf
```

```python
# 데이터 프레임 만들기 3
cdf = pd.DataFrame({"x1" : ["B", "C", "D"],
                    "x2": [2, 3, 4]})
cdf
```

```python
# x1을 key로 해서, 왼쪽(adf)을 기준으로
# 왼쪽의 adf에는 D가 없으므로 해당 값은 NaN으로 변환합니다.
pd.merge(adf, bdf, how='left', on='x1')
```

```python
# x1을 키로 해서 병합, 오른쪽(bdf)을 기준으로
# 오른쪽의 bdf에는 C가 없으므로 해당 값은 NaN으로 변환합니다.
pd.merge(adf, bdf, how='right', on='x1')
```

```python
pd.merge(adf, bdf, how='inner', on='x1')
```

```python
pd.merge(adf, bdf, how='outer', on='x1')
```

```python
# adf와 bdf의 특정한 열을 비교해서 공통 값이 존재하는 경우만 가져옵니다.
# adf.x1열과 bdf.x1열은 A와 B가 같으므로 adf의 해당 값만 출력합니다.
adf[adf.x1.isin(bdf.x1)]
```

```python
adf[~adf.x1.isin(bdf.x1)] # 해당 값만 빼고 출력합니다.
```

```python
pd.merge(adf, cdf)
```

```python
pd.merge(adf, cdf, how='outer')
```

```python
pd.merge(adf, cdf, how='outer', indicator=True)
```

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"')
```

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"').drop(columns=['_merge'])
```
